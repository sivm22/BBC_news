```python
D=A.find_all('span',{'data-testid':"card-metadata-lastupdated"})
print(D)
```

```python
last_updated=[]
```

```python
for l in D:
    last_updated.append(l.get_text())
```

```python
last_updated
```

```python
Timing = []
now = datetime.now() 

for t in last_updated:
  
    match = re.match(r'(\d+)\s+(hrs|hour|minutes|mins|sec|secs|days|day)\s+ago', t)

    if match:
        amount = int(match.group(1))
        unit = match.group(2)

        if 'hr' in unit or 'hour' in unit:
            exact_time = now - timedelta(hours=amount)
        elif 'day' in unit:
            exact_time = now - timedelta(days=amount)
        elif 'min' in unit or 'minutes' in unit:
            exact_time = now - timedelta(minutes=amount)
        elif 'sec' in unit or 'secs' in unit:
            exact_time = now - timedelta(seconds=amount)

        print("Exact timestamp:", exact_time)
        Timing.append(exact_time)
    else:
        
        Timing.append('NA')
        print(f"Could not parse time string.{t}")

for timestamp in Timing:
    print("Formatted timestamp:", timestamp)    
```
